```python
# I am Aadil Nazir 
# Roll_no UE208001, Punjab University...UIET
# I am Working On CPU Bcz I havn't GPU in my  Laptop....
# Distributed Training......
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import logging
import torch.optim.lr_scheduler as lr_scheduler

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

def train():
    try:
        logging.info("Script started...")

        device = torch.device("cpu")
        logging.info("Using CPU for training.")

        logging.info("Preparing dataset...")
        transform = transforms.Compose([
            transforms.RandomHorizontalFlip(),
            transforms.RandomCrop(32, padding=4),
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))
        ])
        dataset = torchvision.datasets.CIFAR10(root='./data', train=True, transform=transform, download=True)
        dataloader = torch.utils.data.DataLoader(dataset, batch_size=32, shuffle=True)
        logging.info("Dataset loaded successfully.")

        logging.info("Initializing model, loss, and optimizer...")
        model = resnet18(num_classes=10).to(device)
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=0.001)

        scheduler = lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)

        # Loop For Traning
        logging.info("Starting training...")
        for epoch in range(5): # Epoch 5
            epoch_loss = 0
            model.train()
            for batch_idx, (inputs, labels) in enumerate(dataloader):
                inputs, labels = inputs.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

                epoch_loss += loss.item()
                if batch_idx % 100 == 0:  # Log every 100 batches
                    logging.info(f"Epoch [{epoch+1}], Batch [{batch_idx+1}/{len(dataloader)}], Loss: {loss.item():.4f}")

            logging.info(f"Epoch [{epoch+1}] completed with average loss: {epoch_loss / len(dataloader):.4f}")
            scheduler.step()  # Step the learning rate scheduler
            # Save Checkpoint
            torch.save({
                'epoch': epoch + 1,
                'model_state': model.state_dict(),
                'optimizer_state': optimizer.state_dict()
            }, f"checkpoint_epoch_{epoch+1}.pth")
            logging.info(f"Checkpoint saved for epoch {epoch+1}.")

        logging.info("Training completed successfully.")

    except Exception as e:
        logging.error(f"An error occurred: {e}", exc_info=True)

if __name__ == "__main__":
    train()
```
